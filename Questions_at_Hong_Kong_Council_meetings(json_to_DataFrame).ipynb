```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib
import json
import re
```

# 1.ViewOralQuestionsEng (Oral Questions)

```python
# Added querty : $orderby=MeetingDate
url = "http://app.legco.gov.hk/QuestionsDB/odata/ViewOralQuestionsEng?$orderby=MeetingDate"     # json is defalt
data = requests.get(url).json()
print(type(data))
```

```python
# We just need the values of 'value')
print(data.keys())
data['value']
```

```python
# Convert Dic to DataFrame
contents_oral=data['value']

Row=[]
MeetingDate=[]
SubjectName=[]
HansardFile=[]
BookmarkName=[]
MemberName=[]
HansardFileURL=[]

for i in contents_oral :
    Row.append(i["Row"])
    MeetingDate.append(i["MeetingDate"])
    SubjectName.append(i["SubjectName"])
    HansardFile.append(i["HansardFile"])
    BookmarkName.append(i["BookmarkName"])
    MemberName.append(i["MemberName"])
    HansardFileURL.append(i["HansardFileURL"])

df_OralQuest=pd.DataFrame({'Row': Row, 
                'MeetingDate': MeetingDate, 
                'SubjectName': SubjectName, 
                'HansardFile': HansardFile, 
                'BookmarkName': BookmarkName, 
                'MemberName': MemberName,
                'HansardFileURL':HansardFileURL})
```

```python
df_OralQuest.to_excel("Oral_Questions.xlsx", index=False)
df_OralQuest
```
