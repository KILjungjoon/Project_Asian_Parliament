## 4.날짜 리스트 생성하기
* 날짜 불러오기

```python
import requests
from bs4 import BeautifulSoup as bs

# url 주소
url="https://www.legco.gov.hk/yr16-17/english/counmtg/question/ques1617.htm#toptbl"

# avoid SSLError: max retries
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
disable_warnings(InsecureRequestWarning)

page = requests.get(url, verify=False)  # SSLError를 피하기 위해 verify=False를 지정
soup = bs(page.text, "html.parser")     # Soup 객체 생성

# Soup 객체를 이용하여 문서를 출력
# print(soup.prettify())
date=[]
for t in soup.select('h2') :
  date.append((t.text).split()[3])
```

* 날짜 리스트 확인

```python
date
```

## 5.컬럼 추가
* for문 반복으로 날짜별 새로운 컬럼 생성

```python
for n in range(4,32) :
  df[n]['term']='6'        # 'term'컬럼 추가
  df[n]['date']=date[(n-4)]    # 'date'컬럼 추가
  df[n]=df[n][['term', 'date', 'Question No.', 'Raised by', 'Subject', 'Reply']]    # 컬럼 순서 변경
```

```python
df[31]    # 컬럼 위치까지 정상
```

## 6.모든 날짜 데이터를 붙이고 저장
* df[4]~df[31]까지 프레임 붙이기

```python
# 먼저 df[4]~df[31]를 리스트에 넣는다.
df_list=[]
for n in range(4, 32) :
  df_list.append(df[n])

# concat으로 붙이기
df_20162017=pd.concat(df_list, ignore_index=True)
# df_20162017

# 저장
df_20162017.to_excel("/content/drive/MyDrive/★Project_prof.Jang/Hong Kong/df_20162017.xlsx", index=False)
```
