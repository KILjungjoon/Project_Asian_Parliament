```python
from nrclex import NRCLex
import nltk
nltk.download('punkt')
text=open('/content/drive/MyDrive/★Project_prof.Jang/Hong Kong/data/clean2020.txt').read()
text_object= NRCLex(text)
```

```python
data = text_object.raw_emotion_scores
data
```

```python
import pandas as pd
emotion_df = pd.DataFrame.from_dict(data, orient='index')
emotion_df = emotion_df.reset_index()
emotion_df = emotion_df.rename(columns={'index' : 'Sentimental Classification' , 0: 'Sentimental Count'})
emotion_df = emotion_df.sort_values(by=['Sentimental Count'], ascending=False)
emotion_df
```

```python
emotion_df['Sentimental Count'].sum()
```

```python
import plotly.express as px
fig = px.bar(emotion_df, x='Sentimental Count', y='Sentimental Classification', color = 'Sentimental Classification',
             title="2020 Sentimental Chart", orientation='h', width = 800, height = 500)
fig.show()
```

```python
# TextBlob Sentiment분석은 좀 모호하다
text2019=open('/content/drive/MyDrive/★Project_prof.Jang/Hong Kong/data/clean2019.txt').read()
text2020=open('/content/drive/MyDrive/★Project_prof.Jang/Hong Kong/data/clean2020.txt').read()
texts=['text2019','text2020']

from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import nltk
nltk.download('movie_reviews')

for text in texts :
  blob = TextBlob(text, analyzer=NaiveBayesAnalyzer())
  print(blob.sentiment)
```
